## What is a Jupyter Notebook?

A **Jupyter notebook** is an interactive document that mixes **explanatory text** (like this) with **live code** and its results. You write and run code in small chunks called *cells*, and the output appears directly below each cell. A notebook is saved as a file with the suffix `.ipynb`.

---

### Why use notebooks?

- **Learn step by step** — Run one cell at a time instead of a whole script, so you can see what each part does and fix things as you go. Great for learning, as you can clearly see the effect of each action.

- **Good for large data** — You can run a subset of cells instead of the whole file (unlike a regular script). Once you load a large dataset in one cell, it stays in memory, so other cells can use it without reloading every time.

- **Experiment safely** — Change a cell and re-run just that cell to try ideas without re-running everything.

- **Visualise the process step by step** — Each cell produces its own output, so you can see the result of every change (e.g. after loading data, after filtering, after aggregating). A script, by contrast, typically has one input and one final output, so the steps in between are hidden.

---

### Using notebooks

#### Adding a cell

- **Add Code** — Top bar, or hover above/below a cell to insert a new cell.

#### Running cells

- **Current cell** — Press the **Play** triangle on the left of the cell.  
  Keeps the state of all previously run cells and **variables** (the names that hold your data and results).

- **All cells** — **Run All** in the top bar.  
  Runs every cell in order and redefines all variables.

- **All above** — **Play** with up arrow (top-right of cell).  
  Runs all cells above the current one; resets variables defined before this cell.

- **Current and below** — **Play** with down arrow (top-right of cell).  
  Runs from the current cell onwards.

- **Restart** — Top bar. Use if you see odd behaviour or bugs; clears the notebook’s memory.

### Filtering

Use square brackets with a **condition** inside: you get back all rows where the condition is true.

- **Single condition** — e.g. keep rows where `currency` equals `'BTC'`:
  ```python
  data_from_csv[data_from_csv['currency'] == 'BTC']
  ```

- **Multiple conditions** — Wrap each condition in parentheses and join with `&` (and) or `|` (or).  
  Example: `(data_from_csv['currency'] == 'BTC') | (data_from_csv['currency'] == 'ETH')`

The cells below show filtering by one value and by a list of values (`.isin()`).

````python
# Equal to value
btc_rows = data_from_csv[data_from_csv['currency'] == 'BTC']
btc_rows
````

````python
# In a list of values
btc_or_eth_rows = data_from_csv[data_from_csv['currency'].isin(['BTC', 'ETH'])]

# Note this is equal to
data_from_csv[(data_from_csv['currency'] == 'BTC') | (data_from_csv['currency'] == 'ETH')]
````

````python
# Negation
# Use ~ to negate a condition (or regular comparative operator)

non_binance_rows = data_from_csv[data_from_csv['exchange'] != 'binance']

non_binance_rows = data_from_csv[~(data_from_csv['exchange'] == 'binance')]
````

## Examples

### Number of deposits to Binance in September 2024

````python
binance_sep_24_deposits = data_from_csv[
    (data_from_csv['exchange'] == 'binance') &
    (data_from_csv['type'] == 'deposit') &
    (data_from_csv['datetime'] >= '2024-09-01') &
    (data_from_csv['datetime'] <= '2024-09-30')
]

binance_sep_24_deposits_count = len(binance_sep_24_deposits)
print(f"Number of deposits to Binance in September 2024: {binance_sep_24_deposits_count}")
binance_sep_24_deposits
````

### Total amount of ETH withdrawn from Binance, Okx, Bybit

````python
eth_withdrawals_from_binance_okx_bybit = data_from_csv[
    (data_from_csv['exchange'].isin(['binance', 'okx', 'bybit'])) &
    (data_from_csv['currency'] == 'ETH') &
    (data_from_csv['type'] == 'withdrawal')
]

total_eth_amount = eth_withdrawals_from_binance_okx_bybit['amount'].sum()

print(f"Total amount of ETH withdrawn from Binance, Okx, Bybit: {total_eth_amount}")
````
